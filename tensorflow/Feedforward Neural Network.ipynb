神經元模型
==========

網路模型以神經元模型為單位組。神經元模型是受人腦神經細胞啟發而設計的數學模型。神經細胞也被稱為神經元(neuron)，神經元模型因此得名。

神經細胞(生物上)
-------------

神經細胞擁有被稱為軸突的管道,軸突可以將電脈衝傳遞到其他神經細胞,然後透過被稱為突觸的接點來告訴下一個神經細胞脈衝的到來。
舉個例子：<br>
**A. 蚊子叮你一個包，你很快就會感覺到痛** <br>
**B. 羽毛輕輕滑過你的皮膚，你根本沒感覺** <br>
如果突觸傳遞強度大，那麼即使只來到一個脈衝，膜電位的變化也很大；反之，如果突觸傳遞強度小，那麼膜電位根本不會有變化。
我們會不斷傳遞的強度的大小，膜電位會積累到一個閥值，輸出脈衝。<br>
意思是我們會不斷受到外界的刺激，當羽毛多到讓你有感覺，就代表你的身體的觸覺系統積累到一個閥值，觸發你有感覺。<br>
可能不是這麼準確的解釋，但我算是可以理解。<br>
<br>
我們人類在學習過程中,大腦內就發生著各種各樣的神經細胞間得突傳遞強度的變化。我們能夠進行各種類型的學習,比如學習語言、自然而然地記住最近發生的事情、經過多次嘗試學會騎自行車等,在每一種類型的學習過程中,大腦內對應的部位都會發生突觸傳遞強度的變化。
